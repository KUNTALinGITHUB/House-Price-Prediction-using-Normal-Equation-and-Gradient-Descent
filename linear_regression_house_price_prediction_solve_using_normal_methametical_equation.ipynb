```python
# ==========================================
# Convert Original Dataset
# -> ML Ready Encoded Dataset
# ==========================================

import pandas as pd


# ==========================================
# Load Dataset
# ==========================================

df = pd.read_csv(
    r"E:\ML\Code_linear_regression\Dataset\House Price Prediction Dataset.csv"
)


# ==========================================
# Convert YearBuilt -> HouseAge
# ==========================================

CURRENT_YEAR = 2026

df["HouseAge"] = CURRENT_YEAR - df["YearBuilt"]

df.drop("YearBuilt", axis=1, inplace=True)


# ==========================================
# Encode Garage
# ==========================================

df["Garage"] = df["Garage"].map({
    "Yes": 1,
    "No": 0
})


# ==========================================
# One-Hot Encode Location
# Reference Category -> Rural
# ==========================================

location_encoded = pd.get_dummies(
    df["Location"],
    prefix="Location",
    dtype=int
)

# Remove Rural manually
location_encoded.drop(
    "Location_Rural",
    axis=1,
    inplace=True
)


# ==========================================
# One-Hot Encode Condition
# Reference Category -> Poor
# ==========================================

condition_encoded = pd.get_dummies(
    df["Condition"],
    prefix="Condition",
    dtype=int
)

# Remove Poor manually
condition_encoded.drop(
    "Condition_Poor",
    axis=1,
    inplace=True
)


# ==========================================
# Remove Original Columns
# ==========================================

df.drop(
    ["Location", "Condition"],
    axis=1,
    inplace=True
)


# ==========================================
# Merge Encoded Columns
# ==========================================

df = pd.concat(
    [df, location_encoded, condition_encoded],
    axis=1
)


# ==========================================
# Reorder Columns
# ==========================================

df = df[
    [
        "Id",

        "Area",
        "Bedrooms",
        "Bathrooms",
        "Floors",
        "HouseAge",

        "Location_Downtown",
        "Location_Suburban",
        "Location_Urban",

        "Condition_Excellent",
        "Condition_Fair",
        "Condition_Good",

        "Garage",

        "Price"
    ]
]


# ==========================================
# Rename Columns
# ==========================================

df.columns = [
    "id",

    "Area(x1)",
    "Bedrooms(x2)",
    "Bathrooms(x3)",
    "Floors(x4)",
    "HouseAge(x5)",

    "Location_Downtown(x6)",
    "Location_Suburban(x7)",
    "Location_Urban(x8)",

    "Condition_Excellent(x9)",
    "Condition_Fair(x10)",
    "Condition_Good(x11)",

    "Garage(x12)",

    "Price(y)"
]


# ==========================================
# Save Encoded Dataset
# ==========================================

output_path = (
    r"E:\ML\Code_linear_regression\Dataset\encoded_house_data.csv"
)

df.to_csv(output_path, index=False)


# ==========================================
# Preview
# ==========================================

print("\nEncoded Dataset Preview:\n")

print(df.head())

print("\n===================================")
print("Encoded CSV Saved Successfully")
print("===================================")
print(f"Location : {output_path}")
```

```python
#  now based on this encoded dataset we can create the input features(x) and target variable(y) matrix .

# x = [1 1360 5 4 3 1970 1 0 0 1 0 0 0] (m x n)# this is the first row of the encoded dataset without the id and price column[bios(b) x1 x2 ...... x14]
# y = [149919                               (n x 1) # this is the first row of the price column which is the target variable we want to predict
#           ]                             

# w = [b
#      w1 
#      w2 
#      w3 
#      w4 
#      w5 
#      w6 
#      w7 
#      w8 
#      w9 
#      w10
#      w11
#      w12 
#      ] (n x 1) # this is the weight vector which will be learned during the training process of the linear regression model

# baseed on this process our entire dataset will be converted into a format that can be used to train a linear regression model to predict house prices based on the input features.

# x = [[1 1360 5 4 3 1970 1 0 0 1 0 0  0],
#      [1 4272 5 4 3 1958 1 0 0 1 0 0 0 ],
#      [1 3592 2 2 3 1938 1 0 0 0 0 1 0 ],
#      ...
#     ]

# y = [149919
#      424998
#      266746
#        ...] so on...

# weigth (w) = (((x.T x)^-1) x.T) y) 
# (n x 1) # this is the normal equation to calculate the weight vector for linear regression model based on the input features and target variable.
```

```python
# =========================================================
# Linear Regression Using Normal Equation
# Stable Version Using Pseudo Inverse
# W = pinv(X) @ y
# =========================================================

import pandas as pd
import numpy as np


# =========================================================
# Load Encoded Dataset
# =========================================================

df = pd.read_csv(
    r"E:\ML\Code_linear_regression\Dataset\encoded_house_data.csv"
)


# =========================================================
# Create Feature Matrix (X)
# Remove:
#   - id column
#   - target column Price(y)
# =========================================================

X = df.drop(columns=["id", "Price(y)"])


# =========================================================
# Convert X into NumPy Array
# =========================================================

X = X.values


# =========================================================
# Add Bias Column
# x0 = 1
# =========================================================

m = X.shape[0]

bias_column = np.ones((m, 1))

X = np.hstack((bias_column, X))


# =========================================================
# Create Target Vector (y)
# =========================================================

y = df["Price(y)"].values.reshape(-1, 1)


# =========================================================
# Calculate Weight Matrix (W)
# Stable Normal Equation
# =========================================================

W = np.linalg.pinv(X) @ y


# =========================================================
# Store Each Weight in Separate Variables
# =========================================================

b   = W[0][0]

w1  = W[1][0]
w2  = W[2][0]
w3  = W[3][0]
w4  = W[4][0]
w5  = W[5][0]

w6  = W[6][0]
w7  = W[7][0]
w8  = W[8][0]

w9  = W[9][0]
w10 = W[10][0]
w11 = W[11][0]

w12 = W[12][0]


```

```python
# =========================================================
# Display Learned Weights
# =========================================================

print("\n=================================================")
print(" Learned Weight Matrix (W)")
print("=================================================\n")

print(f"b   (Bias)                  = {b:.4f}")

print(f"w1  -> Area                 = {w1:.4f}")
print(f"w2  -> Bedrooms             = {w2:.4f}")
print(f"w3  -> Bathrooms            = {w3:.4f}")
print(f"w4  -> Floors               = {w4:.4f}")
print(f"w5  -> HouseAge             = {w5:.4f}")

print(f"w6  -> Location_Downtown    = {w6:.4f}")
print(f"w7  -> Location_Suburban    = {w7:.4f}")
print(f"w8  -> Location_Urban       = {w8:.4f}")

print(f"w9  -> Condition_Excellent  = {w9:.4f}")
print(f"w10 -> Condition_Fair       = {w10:.4f}")
print(f"w11 -> Condition_Good       = {w11:.4f}")

print(f"w12 -> Garage               = {w12:.4f}")


# =========================================================
# Matrix Shapes
# =========================================================

print("\n=================================================")
print(" Matrix Shapes")
print("=================================================\n")

print(f"X shape : {X.shape}")
print(f"y shape : {y.shape}")
print(f"W shape : {W.shape}")


# =========================================================
# Predict All Dataset Values
# y_pred = X @ W
# =========================================================

y_pred = X @ W  # (m x n) @ (n x 1) -> (m x 1)


# =========================================================
# Manual Prediction Formula
# =========================================================
# Example House:
#
# Area        = 2200
# Bedrooms    = 4
# Bathrooms   = 3
# Floors      = 2
# HouseAge    = 10
#
# Downtown    = 1
# Suburban    = 0
# Urban       = 0
#
# Excellent   = 1
# Fair        = 0
# Good        = 0
#
# Garage      = 1
# =========================================================

predicted_price = (
    b
    + (w1 * 2200)
    + (w2 * 4)
    + (w3 * 3)
    + (w4 * 2)
    + (w5 * 10)
    + (w6 * 1)
    + (w7 * 0)
    + (w8 * 0)
    + (w9 * 1)
    + (w10 * 0)
    + (w11 * 0)
    + (w12 * 1)
)


print("\n=================================================")
print(" Manual Prediction")
print("=================================================\n")

print(f"Predicted House Price = {predicted_price:.2f}")
```
