```python
# Gradient Descent

# Starts with random weights and improves them step-by-step.

# Start with bad W
# ↓
# Predict
# ↓
# Measure error
# ↓
# Adjust W slightly
# ↓
# Repeat many times

# until error becomes minimum.
```

```python
# Gradient Descent Pipeline

# Initialize W randomly
#         ↓
# Prediction
#         ↓
# Calculate Error
#         ↓
# Calculate Gradient
#         ↓
# Update Weights
#         ↓
# Repeat
```

```python
# =========================================================
# Linear Regression Using Gradient Descent
# =========================================================

import pandas as pd
import numpy as np


# =========================================================
# Load Dataset
# =========================================================

df = pd.read_csv(
    r"E:\ML\Code_linear_regression\Dataset\encoded_house_data.csv"
)


# =========================================================
# Create Feature Matrix X
# =========================================================

X = df.drop(columns=["id", "Price(y)"]).values


# =========================================================
# Add Bias Column
# =========================================================

m = X.shape[0]

bias = np.ones((m, 1))

X = np.hstack((bias, X))


# =========================================================
# Target Vector y
# =========================================================

y = df["Price(y)"].values.reshape(-1, 1)


# =========================================================
# Feature Scaling (VERY IMPORTANT)
# =========================================================

X_mean = np.mean(X[:, 1:], axis=0)
X_std = np.std(X[:, 1:], axis=0)

X[:, 1:] = (X[:, 1:] - X_mean) / X_std


# =========================================================
# Initialize Weights
# =========================================================

n = X.shape[1]

W = np.zeros((n, 1))


# =========================================================
# Hyperparameters
# =========================================================

learning_rate = 0.01
epochs = 1000


# =========================================================
# Gradient Descent Training
# =========================================================

for epoch in range(epochs):

    # -----------------------------------------
    # Prediction
    # y_hat = XW
    # -----------------------------------------

    y_hat = X @ W


    # -----------------------------------------
    # Error
    # -----------------------------------------

    error = y_hat - y


    # -----------------------------------------
    # Cost Function (MSE)
    # -----------------------------------------

    cost = (1 / m) * np.sum(error ** 2)


    # -----------------------------------------
    # Gradient
    # -----------------------------------------

    gradient = (2 / m) * (X.T @ error)


    # -----------------------------------------
    # Update Weights
    # -----------------------------------------

    W = W - (learning_rate * gradient)


    # -----------------------------------------
    # Print Progress
    # -----------------------------------------

    if epoch % 100 == 0:
        print(f"Epoch {epoch} | Cost = {cost:.2f}")


# =========================================================
# Extract Weights
# =========================================================

b   = W[0][0]

w1  = W[1][0]
w2  = W[2][0]
w3  = W[3][0]
w4  = W[4][0]
w5  = W[5][0]

w6  = W[6][0]
w7  = W[7][0]
w8  = W[8][0]

w9  = W[9][0]
w10 = W[10][0]
w11 = W[11][0]

w12 = W[12][0]


# =========================================================
# Display Final Weights
# =========================================================

print("\n====================================")
print(" Final Learned Weights")
print("====================================\n")

print(f"b   = {b:.4f}")

print(f"w1  = {w1:.4f}")
print(f"w2  = {w2:.4f}")
print(f"w3  = {w3:.4f}")
print(f"w4  = {w4:.4f}")
print(f"w5  = {w5:.4f}")

print(f"w6  = {w6:.4f}")
print(f"w7  = {w7:.4f}")
print(f"w8  = {w8:.4f}")

print(f"w9  = {w9:.4f}")
print(f"w10 = {w10:.4f}")
print(f"w11 = {w11:.4f}")

print(f"w12 = {w12:.4f}")
```

```python
# =========================================================
# Create Test Input
# =========================================================

x_test = np.array([
    [
        1,      # Bias

        2200,   # Area
        4,      # Bedrooms
        3,      # Bathrooms
        2,      # Floors
        10,     # HouseAge

        1,      # Downtown
        0,      # Suburban
        0,      # Urban

        1,      # Excellent
        0,      # Fair
        0,      # Good

        1       # Garage
    ]
])


# =========================================================
# Apply Same Scaling
# =========================================================

x_test[:, 1:] = (
    x_test[:, 1:] - X_mean
) / X_std


# =========================================================
# Predict
# =========================================================

predicted_price = x_test @ W


print("\n====================================")
print(" Test House Prediction")
print("====================================\n")

print(f"Predicted Price = {predicted_price[0][0]:.2f}")
```
